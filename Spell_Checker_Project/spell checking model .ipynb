```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens]
    tokens = [word for word in tokens if word.isalpha()]
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return tokens

file_path = r"C:\Users\Awais\Desktop\NLP TASKs of Jupyter\eng_news_2023_1M\eng_news_2023_1M-sentences.txt"
with open(file_path, "r", errors="replace") as file:
    data = file.readlines()
preprocessed_data = [preprocess_text(line) for line in data]

preprocessed_sentences = [" ".join(tokens) for tokens in preprocessed_data]
for sentence in preprocessed_sentences[:5]:
    print(sentence)
```

## 1st GUI

```python
import tkinter as tk
from tkinter import filedialog, Text
from spello.model import SpellCorrectionModel

sp = SpellCorrectionModel(language='en')
sp.load(r"C:\Users\Awais\Desktop\NLP TASKs of Jupyter\model2.pkl\model.pkl")

root = tk.Tk()
root.title("Spell Correction GUI")
root.geometry("600x600")


input_label = tk.Label(root, text="Enter text:")
input_label.pack()
input_text = tk.Text(root, height=10, width=80)
input_text.pack()
corrected_label = tk.Label(root, text="Corrected text:")
corrected_label.pack()
corrected_text = tk.Text(root, height=10, width=80)
corrected_text.pack()


def spell_correct():
    text = input_text.get("1.0", tk.END).strip()
    correction = sp.spell_correct(text)
    corrected_text.delete("1.0", tk.END)
    corrected_text.insert(tk.END, correction['spell_corrected_text'])

correct_button = tk.Button(root, text="Correct Spelling", command=spell_correct)
correct_button.pack()

root.mainloop()
```

## 2nd GUI

```python
import tkinter as tk
from tkinter import filedialog, Text
from spello.model import SpellCorrectionModel

# Load the trained model
sp = SpellCorrectionModel(language='en')
sp.load(r"C:\Users\Awais\Desktop\NLP TASKs of Jupyter\model2.pkl\model.pkl")
#sp.load(r"C:\Users\Awais\Desktop\NLP TASKs of Jupyter\model2.pkl\en_large.pkl")

root = tk.Tk()
root.title("Spell Correction GUI")
root.geometry("600x600")
root.configure(bg="#000000")  


input_label_font = ("Helvetica", 14, "bold")
input_text_bg = "#333333"  
input_text_fg = "#FFFFFF"  
corrected_text_bg = "#333333"  
corrected_text_fg = "#00FF00"  
button_bg = "#20B2AA"  
button_fg = "#000000"  
label_fg = "#FFFFFF"  


input_label = tk.Label(root, text="Enter text:", font=input_label_font, bg="#000000", fg=label_fg)
input_label.pack(pady=(10, 0))
input_text = tk.Text(root, height=10, width=80, bg=input_text_bg, fg=input_text_fg, font=("Arial", 12))
input_text.pack(pady=(0, 10))
corrected_label = tk.Label(root, text="Corrected text:", font=input_label_font, bg="#000000", fg=label_fg)
corrected_label.pack(pady=(10, 0))
corrected_text = tk.Text(root, height=10, width=80, bg=corrected_text_bg, fg=corrected_text_fg, font=("Arial", 12))
corrected_text.pack(pady=(0, 10))

def spell_correct():
    text = input_text.get("1.0", tk.END).strip()
    correction = sp.spell_correct(text)
    corrected_sentence = correction['spell_corrected_text']
    corrected_text.delete("1.0", tk.END)
    corrected_text.insert(tk.END, corrected_sentence)

correct_button = tk.Button(root, text="Correct Spelling", command=spell_correct, bg=button_bg, fg=button_fg, font=("Helvetica", 14, "bold"))
correct_button.pack(pady=(10, 10))

root.mainloop()
```
